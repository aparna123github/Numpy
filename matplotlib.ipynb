# Matplotlib is Python’s core data visualization library.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import matplotlib.pyplot as plt          # 1. Import matplotlib

x = [1, 2, 3, 4]                          # 2. X data
y = [10, 20, 15, 25]                     # 3. Y data

plt.figure(figsize=(8, 4))               # 4. Set figure size

plt.plot(x, y, label="Sales")            # 5. Draw line plot

plt.xlabel("Month")                      # 6. Label X-axis
plt.ylabel("Revenue")                    # 7. Label Y-axis

plt.title("Monthly Sales Report")        # 8. Add title

plt.legend()                             # 9. Show legend

plt.grid(True)                           # 10. Add grid

plt.show()                               # 11. Display plot
```

### LINE PLOT (TREND ANALYSIS)

```python
# Data
days = np.arange(1, 8)
sales = [120, 150, 170, 160, 180, 200, 220]

# Plot
plt.plot(days, sales)
plt.xlabel("Day")
plt.ylabel("Sales")
plt.title("Weekly Sales Trend")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Scatter Plot Example
x = [1, 2, 3, 4, 5]
y = [5, 15, 10, 20, 25]

plt.figure(figsize=(7,5))
plt.scatter(x, y, color='purple', marker='o')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Demo Scatter Plot")
plt.grid(True) #this is the reason we r having grid lines
plt.show()
```

## HISTOGRAM (DISTRIBUTION OF DATA)

```python
ages = [22, 25, 27, 29, 30, 31, 35, 40, 45, 50, 60]

plt.hist(ages, bins=5)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

```python
data = [12, 15, 12, 18, 19, 12, 17, 20, 15, 18]

plt.figure(figsize=(6,4))
plt.hist(data, bins=5, color='green', edgecolor='black')
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.title("Demo Histogram")
plt.grid(axis='y')
plt.show()
```

## Bar Chart (Category Comparison)

```python
products = ["A", "B", "C", "D"]
revenue = [400, 700, 300, 900]

plt.bar(products, revenue)
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.title("Revenue by Product")
plt.show()
```

### HORIZONTAL BAR CHART

```python
categories = ['A', 'B', 'C', 'D']
values = [12, 18, 7, 15]

plt.figure(figsize=(7,4))
plt.barh(categories, values, color='skyblue') #barh IS THE REASON OF HORIZONTAL BARCHART
plt.xlabel("Value")
plt.ylabel("Category")
plt.title("Demo Horizontal Bar Chart")
plt.grid(axis='x')
plt.show()
```

## Subplots (Real Data Science Workflow)

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# Line plot
ax[0].plot(days, sales)
ax[0].set_title("Sales Trend")

# Histogram
ax[1].hist(sales, bins=5)
ax[1].set_title("Sales Distribution")

plt.show()
```

## Multiple Line Plot with Legend

```python
months = [1, 2, 3, 4]
product_a = [10, 20, 15, 25]
product_b = [5, 15, 10, 20]

plt.figure(figsize=(8,5))
plt.plot(months, product_a, label="Product A", marker='o')
plt.plot(months, product_b, label="Product B", marker='s')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Comparison")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

days = np.arange(1, 8)
sales = np.array([120, 150, 170, 160, 180, 200, 220])

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].plot(days, sales, marker='o', linestyle='--')
ax[0].set_title("Sales Trend Over Time")
ax[0].set_xlabel("Day")
ax[0].set_ylabel("Sales")

ax[1].hist(sales, bins=5)
ax[1].set_title("Sales Distribution")
ax[1].set_xlabel("Sales Value")
ax[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

### practice 2

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)  # for reproducibility

group_a = np.random.normal(70, 10, 1000)
group_b = np.random.normal(75, 12, 1000)

plt.figure(figsize=(8,4))
plt.hist(group_a, bins=30, alpha=0.6, label='Group A')
plt.hist(group_b, bins=30, alpha=0.6, label='Group B')
plt.title("Score Distribution Comparison")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
#no. of bins increased 
# finer details will be there
plt.hist(group_a, bins=60, alpha=0.6)
plt.hist(group_b, bins=60, alpha=0.6)
plt.show()
```

```python
# copmpares distribution shape
plt.hist(group_a, bins=30, density=True, alpha=0.6, label='Group A')
plt.hist(group_b, bins=30, density=True, alpha=0.6, label='Group B')
plt.legend()
plt.show()
```

```python
# how to show them without overlap
plt.hist([group_a, group_b], bins=30, stacked=True, label=['A','B'])
plt.legend()
plt.show()
#stacked = true, tells the graph to DO NOT OVERLAP BARS,instead ACCUMULATE THEM
```

```python
#ahow the mean value on same histo
plt.hist(group_a, bins=30, alpha=0.6)
plt.hist(group_b, bins=30, alpha=0.6)

plt.axvline(np.mean(group_a), linestyle='--')
plt.axvline(np.mean(group_b), linestyle='--')

plt.show()
```
