## Statisctical functiom

### NumPy statistical functions reduce data into meaningful numbers by summing, averaging, or 
measuring spread — either across rows, columns, or the entire array.

```python
a = np.array([[1,2,3],[4,5,6]])
print(a)
```

```python
print(np.sum(a))
print(np.sum(a, axis=1)) #ROW WISE SUM
print(np.sum(a, axis=0)) #COL WISE SUM
```

```python
print(np.mean(a))
print(np.mean(a,axis = 1)) #row wise mean
print(np.mean(a,axis = 0)) #col wise mean
```

```python
print(np.median(a))
```

```python
print(np.min(a))
print(np.min(a, axis=0))
print(np.max(a))
print(np.max(a, axis=0))
```

```python
np.ptp(a)
#range = max - min
```

```python
print(np.median(a))
print(np.std(a)) #std. deviation
print(np.cumsum(a)) #cummalative son
print(np.argmin(a)) #index of min
print(np.argmax(a)) #index of max
```

## Broadcasting

```python
a = np.array([1, 2, 3])
b = np.array([10])

a + b
# 10 is added to all the col
#b has one element → reused for every element in a
```

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([10, 20, 30])

a + b
#here every element of b is added to a
```

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])
c = np.array([[10],
              [20]])

a + c
#col wise broadcasting
```
