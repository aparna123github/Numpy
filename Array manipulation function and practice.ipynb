## Set operation

```python
# find unique elements
a = np.array([1, 2, 2, 3, 4, 4])
b = np.unique(a)
print(b)
```

```python
#differnce of a and b
a = np.array([1, 2, 3, 4])
b = np.array([3, 4])

c = np.setdiff1d(a, b)
print(c)
```

```python
#intersection of two array
set_a = np.array([1,2,3])
set_b = np.array([1,2,3])
intersection = np.intersect1d(set_a, set_b)
print("intersection of a and b:", intersection)
```

```python
#union of two arrays
union = np.union1d(set_a, set_b)
print(union)
```

```python
# MEMBERSHIP TEST
a = np.array([1, 2, 3, 4])
b = np.array([2, 4])

c = np.isin(a, b)
print(c)
```

```python
#improved membership test
a = np.array([10, 20, 30, 40])
b = np.array([20, 40])

c = np.isin(a, b)
print(c)
```

```python
#works element wise in array
a = np.array([1, 2, 3, 4, 5])
print(np.isin(3, a))        # True
print(np.isin([2, 7], a))   # [True False]
```
