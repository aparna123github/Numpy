```python
import numpy as np

Students = ["A","B","C","D"]
Subjects = ["Math","Science","English"]

# Marks
Marks = np.array([
    [85, 78, 92],   # A
    [88, 90, 79],   # B
    [70, 75, 80],   # C
    [95, 92, 88]    # D
])
```

```python
# Find average marks of each student
avg_student = Marks.mean(axis = 1)
avg_student
```

```python
# Find subject wiese average
avg_sub = Marks.mean(axis = 0)
avg_sub
```

```python
# Who scored highest in maths
math_topper = Students[np.argmax(Marks[:,0])] #ALL ROWS FIRST COL
math_topper 
```

```python
# Total marks of student c
marks = np.sum(Marks[2])
marks
```

```python
#Highest marks in each student
highest = np.max(Marks,axis = 0)
highest
# np.max must receive an ARRAY as its first argument.(Marks)
```

```python
# Lowest marks in each student
lowest = np.min(Marks,axis = 1)
lowest
```

```python
# Total marks of each student
total = np.sum(Marks,axis = 1)
total
```

```python
axis=0 → max column-wise

axis=1 → max row-wise

no axis → max of entire array
```

```python
# Student scoring more than 80 in maths
len(Marks[Marks[:,0] > 80])
```

```python
# Marks of student b
Marks[Students.index("B")]
```

```python
# Subject with highest average
Subjects[np.argmax(Marks.mean(axis=0))]
```

```python
# Subject with lowest average
student_avg = Marks.mean(axis=1)
student_avg
Students[np.argmin(student_avg)]
```

```python
# Second highest score
sorted(total)[-2]
```

# Example: Employee Performance Data

```python
# employee_performance.py
import numpy as np

# Employees and Departments
Employees = ["John", "Alice", "Bob", "Diana"]
Departments = ["Sales", "Marketing", "HR"]

# Performance scores (out of 100)
Scores = np.array([
    [80, 90, 70],   # John
    [85, 95, 75],   # Alice
    [60, 70, 65],   # Bob
    [90, 85, 80]    # Diana
])
```

```python
# Average score of each employe
score = Scores.mean(axis = 1)
score
```

```python
# Average score per department
score = Scores.mean(axis = 0)
score
```

```python
# Employe with highest total score
highest = Scores.sum(axis = 1)
np_max = np.argmax(highest)
highest
```

```python
# 3. Employee with highest total score
total_scores = Scores.sum(axis=1)
top_employee = Employees[np.argmax(total_scores)]
print("Top employee:", top_employee)
```
